```python
import matplotlib as mpl
import matplotlib.pyplot as plt
#import seaborn as sns
mpl.style.use('ggplot')
```

```python
import pandas as pd
ruta="/workspaces/Analisis-con-python/"
tabla=pd.read_csv( ruta + 'fra.csv',sep=',',encoding='utf-8')
tabla
```

```python
tabla.info()
```

```python
tabla.isnull().sum()
```

```python
SEXO_1=tabla['SEXO'].value_counts()
SEXO_1
```

```python
SEXO_1=tabla['SEXO'].value_counts().index
SEXO_1
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.bar(SEXO_1, tabla['SEXO'].value_counts(),color='c')
ax.set_title('Perfil de Victima')
ax.set_ylabel('Cantidad de personas')
ax.set_xlabel('SEXO')
#ax.set_xticklabels(SEXO_1, rotation=90)
#ax.text(6-0.3,16500,'Hi',fontsize=8,color='red',fontweight='bold')
```
